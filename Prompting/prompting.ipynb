```python
import os 
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage, AIMessage
from langchain_groq import ChatGroq
from pydantic import BaseModel
```

```python
api_key = os.getenv("GROQ_API_KEY")
model = ChatGroq(
    model="meta-llama/llama-4-scout-17b-16e-instruct",
    api_key=api_key
)
```

```python
agent = create_agent(model=model)
```

```python
question = HumanMessage(content="Which is the state food of maharashtra?")

response = agent.invoke(
    {"messages": [question]}
)

print(response['messages'][1].content)
```

```python
system_prompt = "You are a food historian and Food historian is a person who studies the history of food, recipes, eating habits, and how cultures have used food over time."

food_historian = create_agent(
    model=model,
    system_prompt=system_prompt
)

response = food_historian.invoke(
    {"messages": [question]}
)

print(response['messages'][1].content)
```

```python
system_prompt = """

You are a food historian and Food historian is a person who studies the history of food, recipes, eating habits, and how cultures have used food over time.

User: Which is the food which similar to the burger?
Food Historian: Vada Pav

User: Which are the dishes of maharashtra uses pav with the cuisine?
Food Historian: Misal Pav, Pav Bhaji

"""

question = "Which are the types on chapati's are there in maharashtra?"

food_historian = create_agent(
    model=model,
    system_prompt=system_prompt
)

response = food_historian.invoke(
    {"messages": [question]}
)

print(response['messages'][1].content)
```
